```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

spec_many_df = pd.read_csv('spectrum_many_180.csv')
spec_many_df1 = pd.read_csv('spectrum_many_180_1.csv')
spec_one_df = pd.read_csv('spectrum_one_180.csv')
spec_one_df1 = pd.read_csv('spectrum_one_180_1.csv')

spec_many_ch = spec_many_df.values[:, 0]
spec_many_counts = spec_many_df.values[:, 1]
spec_many_ch1 = spec_many_df1.values[:, 0]
spec_many_counts1 = spec_many_df1.values[:, 1]
spec_one_ch = spec_one_df.values[:, 0]
spec_one_counts = spec_one_df.values[:, 1]
spec_one_ch1 = spec_one_df1.values[:, 0]
spec_one_counts1 = spec_one_df1.values[:, 1]

plt.figure(figsize=(10, 6))
plt.step(spec_many_ch, spec_many_counts, label='50s (1)')
plt.step(spec_many_ch1, spec_many_counts1, label='50s (2)')
plt.step(spec_one_ch, spec_one_counts, label=r'$10 \times 5$s (1)')
plt.step(spec_one_ch1, spec_one_counts1, label=r'$10 \times 5$s (2)')
plt.xlabel('Channel')
plt.ylabel('Counts')
plt.legend()
plt.yscale('log')
plt.savefig('one_many_test.pdf')
plt.show()
```
