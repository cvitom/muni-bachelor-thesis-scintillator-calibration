```python
import numpy as np
import matplotlib.pyplot as plt
import pandas  as pd

# list files in pid/ directory
import os
import glob
import re

# get the list of files
files = glob.glob("pid/*.dat")
# get file names only
filenames = [os.path.basename(f) for f in files]
print(files)

for i, f in enumerate(files):
    # read the data
    df = pd.read_csv(f, sep=' ', header=None)
    temp = df.values[:, 0]
    power = df.values[:, 1]
    # plot the data
    plt.figure()
    ax1 = plt.gca()
    ax2 = plt.twinx()
    ax1.plot(np.arange(len(temp)), temp, 'r', label='Temperature')
    ax2.plot(np.arange(len(temp)), power, 'b', label='Power')
    plt.legend()
    plt.title(f)
    plt.show()
```
